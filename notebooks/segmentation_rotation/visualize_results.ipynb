# Visualización de Resultados de Entrenamiento
## Comparación de Modelos: MobilNet, ResNet50 y EfficientNet

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np

# Configuración de estilo
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## Cargar los datos de los 3 modelos

```python
# Ruta a los datos
data_path = 'Data_results'
images_path = 'images'

# Crear carpeta de imágenes si no existe
os.makedirs(images_path, exist_ok=True)
print(f"Carpeta '{images_path}' creada/verificada")

# Diccionario con los modelos
models = {
    'MobilNet': 'MobilnetSegRot_20251103-164956',
    'ResNet50': 'ResNet50SegRot_20251103-174152',
    'EfficientNet': 'EfficientNetSegRot_20251103-193339'
}

# Métricas a visualizar
metrics = ['f1', 'mean_iou', 'loss']
splits = ['train', 'val']

# Cargar todos los datos
data = {}
training_info = {}

for model_name, model_id in models.items():
    data[model_name] = {}
    training_info[model_name] = {}
    
    for metric in metrics:
        data[model_name][metric] = {}
        for split in splits:
            filename = f'run-{model_id}-tag-{metric}_{split}.csv'
            filepath = os.path.join(data_path, filename)
            df = pd.read_csv(filepath)
            data[model_name][metric][split] = df
    
    # Calcular información de entrenamiento
    train_df = data[model_name]['f1']['train']
    training_info[model_name]['epochs'] = len(train_df)
    training_info[model_name]['total_time_seconds'] = train_df['Wall time'].iloc[-1] - train_df['Wall time'].iloc[0]
    training_info[model_name]['total_time_hours'] = training_info[model_name]['total_time_seconds'] / 3600
    training_info[model_name]['time_per_epoch'] = training_info[model_name]['total_time_seconds'] / training_info[model_name]['epochs']

print("Datos cargados exitosamente")
print("\n" + "="*80)
print("INFORMACIÓN DE ENTRENAMIENTO")
print("="*80)
for model_name, info in training_info.items():
    print(f"\n{model_name}:")
    print(f"  - Épocas: {info['epochs']}")
    print(f"  - Tiempo total: {info['total_time_hours']:.2f} horas ({info['total_time_seconds']:.0f} segundos)")
    print(f"  - Tiempo por época: {info['time_per_epoch']:.2f} segundos")
print("="*80)
```

## 1. F1 Score - Entrenamiento (Train)

```python
plt.figure(figsize=(12, 6))

colors = {'MobilNet': 'blue', 'ResNet50': 'green', 'EfficientNet': 'red'}

for model_name in models.keys():
    df = data[model_name]['f1']['train']
    plt.plot(df['Step'], df['Value'], label=model_name, color=colors[model_name], linewidth=2)

plt.xlabel('Epoch', fontsize=22)
plt.ylabel('F1 Score', fontsize=22)
plt.title('F1 Score - Entrenamiento (Train)', fontsize=24, fontweight='bold')
plt.legend(fontsize=20)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(os.path.join(images_path, '01_f1_train.png'), dpi=300, bbox_inches='tight')
plt.show()
print("Guardado: 01_f1_train.png")
```

## 7. Comparación de Todas las Métricas (2x3 Grid)

```python
fig, axes = plt.subplots(3, 2, figsize=(16, 12))
fig.suptitle('Comparación Completa de Modelos', fontsize=16, fontweight='bold')

# F1 Train
for model_name in models.keys():
    df = data[model_name]['f1']['train']
    axes[0, 0].plot(df['Step'], df['Value'], label=model_name, color=colors[model_name], linewidth=2)
axes[0, 0].set_title('F1 Score - Train', fontweight='bold')
axes[0, 0].set_xlabel('Epoch')
axes[0, 0].set_ylabel('F1 Score')
axes[0, 0].legend(fontsize=14, title_fontsize=16)
axes[0, 0].grid(True, alpha=0.3)

# F1 Val
for model_name in models.keys():
    df = data[model_name]['f1']['val']
    axes[0, 1].plot(df['Step'], df['Value'], label=model_name, color=colors[model_name], linewidth=2)
axes[0, 1].set_title('F1 Score - Val', fontweight='bold')
axes[0, 1].set_xlabel('Epoch')
axes[0, 1].set_ylabel('F1 Score')
axes[0, 1].legend(fontsize=14, title_fontsize=16)
axes[0, 1].grid(True, alpha=0.3)

# IoU Train
for model_name in models.keys():
    df = data[model_name]['mean_iou']['train']
    axes[1, 0].plot(df['Step'], df['Value'], label=model_name, color=colors[model_name], linewidth=2)
axes[1, 0].set_title('Mean IoU - Train', fontweight='bold')
axes[1, 0].set_xlabel('Epoch')
axes[1, 0].set_ylabel('Mean IoU')
axes[1, 0].legend(fontsize=14, title_fontsize=16)
axes[1, 0].grid(True, alpha=0.3)

# IoU Val
for model_name in models.keys():
    df = data[model_name]['mean_iou']['val']
    axes[1, 1].plot(df['Step'], df['Value'], label=model_name, color=colors[model_name], linewidth=2)
axes[1, 1].set_title('Mean IoU - Val', fontweight='bold')
axes[1, 1].set_xlabel('Epoch')
axes[1, 1].set_ylabel('Mean IoU')
axes[1, 1].legend()
axes[1, 1].grid(True, alpha=0.3)

# Loss Train
for model_name in models.keys():
    df = data[model_name]['loss']['train']
    axes[2, 0].plot(df['Step'], df['Value'], label=model_name, color=colors[model_name], linewidth=2)
axes[2, 0].set_title('Loss - Train', fontweight='bold')
axes[2, 0].set_xlabel('Epoch')
axes[2, 0].set_ylabel('Loss')
axes[2, 0].legend()
axes[2, 0].grid(True, alpha=0.3)

# Loss Val
for model_name in models.keys():
    df = data[model_name]['loss']['val']
    axes[2, 1].plot(df['Step'], df['Value'], label=model_name, color=colors[model_name], linewidth=2)
axes[2, 1].set_title('Loss - Val', fontweight='bold')
axes[2, 1].set_xlabel('Epoch')
axes[2, 1].set_ylabel('Loss')
axes[2, 1].legend()
axes[2, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(images_path, '07_comparacion_completa.png'), dpi=300, bbox_inches='tight')
plt.show()
print("Guardado: 07_comparacion_completa.png")
```

## 8. Tabla Resumen - Mejores Valores

```python
# Crear tabla resumen con los mejores valores de cada modelo
summary_data = []

for model_name in models.keys():
    row = {'Modelo': model_name}
    
    # Información de entrenamiento
    row['Épocas'] = training_info[model_name]['epochs']
    row['Tiempo Total (h)'] = round(training_info[model_name]['total_time_hours'], 2)
    row['Tiempo/Época (s)'] = round(training_info[model_name]['time_per_epoch'], 2)
    
    # F1 Score
    row['F1 Train (max)'] = round(data[model_name]['f1']['train']['Value'].max(), 4)
    row['F1 Val (max)'] = round(data[model_name]['f1']['val']['Value'].max(), 4)
    
    # IoU
    row['IoU Train (max)'] = round(data[model_name]['mean_iou']['train']['Value'].max(), 4)
    row['IoU Val (max)'] = round(data[model_name]['mean_iou']['val']['Value'].max(), 4)
    
    # Loss
    row['Loss Train (min)'] = round(data[model_name]['loss']['train']['Value'].min(), 4)
    row['Loss Val (min)'] = round(data[model_name]['loss']['val']['Value'].min(), 4)
    
    summary_data.append(row)

summary_df = pd.DataFrame(summary_data)
print("\n" + "="*100)
print("RESUMEN COMPLETO - MÉTRICAS Y TIEMPO DE ENTRENAMIENTO")
print("="*100)
print(summary_df.to_string(index=False))
print("="*100)
```

## 9. Gráfico de Barras Comparativo - Mejores Valores

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Comparación de Mejores Valores por Modelo', fontsize=16, fontweight='bold')

x = np.arange(len(models))
width = 0.35
model_names = list(models.keys())

# F1 Score
f1_train = [data[m]['f1']['train']['Value'].max() for m in model_names]
f1_val = [data[m]['f1']['val']['Value'].max() for m in model_names]
axes[0].bar(x - width/2, f1_train, width, label='Train', alpha=0.8)
axes[0].bar(x + width/2, f1_val, width, label='Val', alpha=0.8)
axes[0].set_xlabel('Modelo')
axes[0].set_ylabel('F1 Score')
axes[0].set_title('F1 Score Máximo', fontweight='bold')
axes[0].set_xticks(x)
axes[0].set_xticklabels(model_names, rotation=15)
axes[0].legend()
axes[0].grid(True, alpha=0.3, axis='y')

# IoU
iou_train = [data[m]['mean_iou']['train']['Value'].max() for m in model_names]
iou_val = [data[m]['mean_iou']['val']['Value'].max() for m in model_names]
axes[1].bar(x - width/2, iou_train, width, label='Train', alpha=0.8)
axes[1].bar(x + width/2, iou_val, width, label='Val', alpha=0.8)
axes[1].set_xlabel('Modelo')
axes[1].set_ylabel('Mean IoU')
axes[1].set_title('Mean IoU Máximo', fontweight='bold')
axes[1].set_xticks(x)
axes[1].set_xticklabels(model_names, rotation=15)
axes[1].legend()
axes[1].grid(True, alpha=0.3, axis='y')

# Loss
loss_train = [data[m]['loss']['train']['Value'].min() for m in model_names]
loss_val = [data[m]['loss']['val']['Value'].min() for m in model_names]
axes[2].bar(x - width/2, loss_train, width, label='Train', alpha=0.8)
axes[2].bar(x + width/2, loss_val, width, label='Val', alpha=0.8)
axes[2].set_xlabel('Modelo')
axes[2].set_ylabel('Loss')
axes[2].set_title('Loss Mínimo', fontweight='bold')
axes[2].set_xticks(x)
axes[2].set_xticklabels(model_names, rotation=15)
axes[2].legend()
axes[2].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig(os.path.join(images_path, '08_barras_comparativas.png'), dpi=300, bbox_inches='tight')
plt.show()
print("Guardado: 08_barras_comparativas.png")
```
